```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import locale
```

```python
df=pd.read_csv("input/pc21.csv")
```

```python
df.drop(['Contract ', 'Sales Order', 'Product: Product Name'],1,inplace=True)
```

```python
df.head()
```

```python
df = df.rename(columns={'Created Date':'Date', 'Product: Metal: Metal Name':'Metal', 'Quantity Ordered':'Quantity','Customer ':'Customer', 'Product: Product Code':'Code'})
```

```python
df.head()
```

```python
df = df[['Date','Customer','Metal','Code','Quantity', 'Unit Price']]
```

```python
df['Total Price'] = df['Quantity'] * df['Unit Price']
```

```python
df.head()
```

```python
locale.setlocale(locale.LC_ALL, '')
```

```python
def curr(value):
    return locale.currency(value, grouping=True)
```

```python
df['Unit Price'] = df['Unit Price'].map(curr)
```

```python
df['Total Price'] = df['Total Price'].map(curr)
```
